```python
import random
import matplotlib
import matplotlib.pyplot as plt

A=[]
B=[]
w_x,w_y=1,1
bias=5
l=0.25
n=0
        
for i in range(1000):
    x=random.uniform(0,100)
    A.append(x)
    y=random.uniform(-100,100)
    B.append(y)

def check(v):
    if v>0:
        return 1
    else:
        return 0
    
def activate(v,u,e):
    global w_x,w_y,l
    w_x=w_x+v*e*l
    w_y=w_y+u*e*l

def feed_forward(v,u):
    global bias,l,w_x,w_y,n
    y_axis=check(u)
    ff=(w_x*v+w_y*u+bias)
    h=check(ff)
    if h>0:
        plt.plot(v,u,'r+')
    else:
        plt.plot(v,u,'b+')
    error=h-y_axis
    if error!=0:
        activate(v,u,error)
        print(w_x,w_y)

for i in range(len(A)):
    feed_forward(A[i],B[i])
    n=n+1
plt.show()

    
    
    

    
```
